##### Clean dataset
- I would be cleaning the dataset in 4 steps

- Cleaning Step 1 - I will be dropping any column i consider unnecessary

```python
df.columns
```

```python
df = df.drop(columns= (['ListingKey', 'ListingNumber', 'LoanKey', 'MemberKey', 'GroupKey', 'LoanOriginationQuarter', 'ClosedDate', 'Recommendations', 'InvestmentFromFriendsCount', 'InvestmentFromFriendsAmount', 'LP_CollectionFees', 'OnTimeProsperPayments', 'ProsperPaymentsLessThanOneMonthLate', 'ProsperPaymentsOneMonthPlusLate', 'ProsperPrincipalBorrowed', 'ProsperPrincipalOutstanding', 'ScorexChangeAtTimeOfListing']))
```

```python
df.info()
```

- Cleaning step 2- input missing values for numerical columns with the average

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())
```

```python
df.info()
```

- Cleaning step 3 - fill all empty categorical columns with mode

```python
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
df.info()
```

- Cleaning Step 4 - Convert date columns to datetime format

```python
date_columns = ['ListingCreationDate', 'DateCreditPulled', 'LoanOriginationDate']
df[date_columns] = df[date_columns].apply(pd.to_datetime, errors='coerce')
```

```python
df.info()
```

#### Question 6 
- What is the correlation between loan status and loan term?

```python
df.groupby('LoanStatus')['Term'].value_counts(normalize=True).unstack().plot(
    kind='bar', 
    stacked=True, 
    figsize=(10, 6), 
    title='Loan Status by Term', 
    colormap='viridis')
plt.xlabel('Loan Status')
plt.ylabel('Proportion')
plt.legend(title='Loan Term')
plt.show()
```

#### Observation
- This chart gives a detailed look at how the different loans by term that is; 12 months, 36 months and 60 months correlate with their current status for example current, completed, defaulted etc.
- when we look at the loan performance across the terms we can see that;
- for the 36 months loans the majority are cancelled, completed and defaulted
- for the 60 months loans, the majority are still current and this may be because there is an extended payment period, although the completed loans are fewer.
- for the 12 months loans, the majority are completed and this could be because ther term is shorter and so there is a quicker closure period. we can also see that there are no defaulted loans compared to the other terms which also shows a lower time frame and risk of repayment.

#### Question 7 
- What is the relationship between debt-to-income ratio and loan amount?

```python
plt.scatter(df['DebtToIncomeRatio'], df['LoanOriginalAmount'], color='teal', alpha=0.5)
plt.title('Debt-to-Income Ratio vs Loan Amount')
plt.xlabel('Debt-to-Income Ratio')
plt.ylabel('Loan Original Amount')
plt.show()
```

#### Observation
- here we can observe from the chart that borrowers with higher loan amounts have lower debt-to-income ratios, this means that the instituition prefers to give larger loans to borrowers that have manageable debt relative to their income, this is because they are seen as less risky as their income can cover new and existing debts
- we can also see that borrowers with higher debt-to-income ratios tend to have smaller loan amounts, this is mostly because higher debt-to-income ratio is risky
-  there are also some outliers, as we can see. there are some borrowers with high loan amounts and high debt-to-income ratios, these cases may be because of great credit history or other income sources that do not have ab influence on the debt-to-income ratio

#### Question 9
- What is the average borrower APR by employment status?

```python
avg_apr_by_employment = df.groupby('EmploymentStatus')['BorrowerAPR'].mean()
plt.figure(figsize=(10, 6))
avg_apr_by_employment.sort_values().plot(kind='bar', color='teal')
plt.title('Average Borrower APR by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Average Borrower APR')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Observation
- This chart shows the average borrower APR(Annual Percentage Rate) across the different employment statuses and also shows how employment profiles and job stability could have an impact on lending terms
- we can see that employed borrowers benefit from the lowest average APR and this shows that employment is a strong indicator of creditworthiness.
- it also shows that the APRs for self-employed people are slightly higher than those of employed people, this maybe because entrepreneurship is perceived as being less stable
- in addition, we see that unemployed people have the highest APRs, the lack of income increases risks and this leads to higher rates. also retired and part-time people have moderate APRs.

#### Question 10
- what is the average loan amount by prosper rating?

```python
avg_loan_by_rating = df.groupby('ProsperRating (Alpha)')['LoanOriginalAmount'].mean()
plt.figure(figsize=(10, 6))
avg_loan_by_rating.sort_index().plot(kind='bar', color='teal')
plt.title('Average Loan Amount by Prosper Rating')
plt.xlabel('Prosper Rating (Alpha)')
plt.ylabel('Average Loan Original Amount')
plt.tight_layout()
plt.show()
```

#### Observation
- this chart shows the relationship between the average loan original amount and prosper rating (alpha), it gives an insight into how creditworthiness shown by prosper rating impacts loan amounts
- we can see that the higher prosper ratings include;
- high rating AA, A
- moderate rating B, C
- low rating D, E, HR
- borrowers in the AA and A categories receive the highest average loan amounts and this shows excellent creditworthiness and low risk for repayment or default
- borrowers in B and C category also receive high to moderate loan amounts which means they have good financial profiles
- borrowers in the D, E and HR categories have lower rating and as such have access to significantly lower loan amounts as a result of higher risks.
- we also observed that some borrowers with lower prosper ratings received high loan amounts, this could be as a result of other compensating factors like higher income, collateral and so on. 

## Conclusions

#### Key Findings
- Borrower APR Distribution:
The distribution of Borrower APR is right-skewed, which shows that most borrowers are charged moderate to high APRs, with fewer benefiting from lower rates. Borrowers with better Prosper Ratings enjoy more favorable APRs.

- Income Range Distribution:
A majority of borrowers fall into mid-income brackets. Borrowers with the lowest or unreported income ranges are less represented, which shows limited access to loans and lending practices for these groups.

- Employment Status:
Borrowers who are employed full-time constitute the majority, indicating that lenders prefer extending credit to borrowers with stable income sources. Unemployed and retired individuals have lower representation.


- Homeownership:
A higher percentage of borrowers are not homeowners, this suggests that Prosper loans might cater to individuals looking for financial flexibility rather than property-backed credit options.

- Loan Terms:
The 36-month term is the most common, followed by 60-month loans. Shorter terms may appeal to borrowers seeking quicker repayment, while longer terms cater to larger financial needs.

- Loan Status by Term:
Most loans are either current or fully paid, this shows a healthy loan repayment behavior. Longer-term loans tend to have a higher proportion of defaults in repayment, reflecting the added risk of extended repayment periods.

- Debt-to-Income Ratio and Loan Amount:
Borrowers with moderate debt-to-income ratios tend to secure larger loans. Extremely high ratios limit borrowing capacity, showing lender caution.

- Credit Score and Loan Amount:
Higher credit scores have a relationship with larger loan amounts, this highlights the importance of strong credit profiles in accessing greater financial resources.

- Borrower APR and Employment Status:
Borrowers with stable employment (full-time, part-time) receive lower APRs compared to unemployed borrowers, reflecting lender confidence in income stability.

- Loan Original Amount and Prosper Rating:
Borrowers with better Prosper Ratings secure significantly larger loans, this shows the role of creditworthiness in loan approvals.

#### Insights
- Creditworthiness Matters: Features like Prosper Ratings, credit scores, and debt-to-income ratios consistently show a strong influence on loan terms, amounts, and APRs.
- Stable Income is Key: Employment status plays an important role in both loan access and affordability, with full-time employed borrowers havng the best benefits.
